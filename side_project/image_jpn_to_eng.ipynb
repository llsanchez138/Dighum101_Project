```python
#openning and reading all images in folder: shounen
## remove .DS_Store files in bash: rm -v **/.DS_Store 

indir = r'shounen/'
text_list = []
for root, dirs, filenames in os.walk(indir):
    for filename in filenames:
        im = Image.open(indir + filename)
        all_img_text = pytesseract.image_to_string(im, lang='eng').lower()
        text_list = text_list + all_img_text.split()
```

```python
def remove_punctuation(word_list):
    for i in range(len(word_list)):
        word = word_list[i]
        for letter in word:
             if (letter in string.punctuation):
                word = word.replace(letter, "")
        word_list[i] = word
    return word_list
      
text_list = remove_punctuation(text_list)  
print(text_list)
```

```python
#joining strings from shounen list to create tokens
join_text = " ".join(text_list).replace("wangareadernet", "").replace("fpp", "")
join_tokens = join_text.split()

#removing stopwords from shounen tokens
no_stops = [word for word in join_tokens if word not in stopwords.words('english')]
no_stops
```

```python
#frequency of common words in shounen
freq2 = Counter(no_stops)
freq2.most_common(10)
```
