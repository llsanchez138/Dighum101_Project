```python
import string

def remove_punctuation(word_list):
    for i in range(len(word_list)):
        word = word_list[i]
        for letter in word:
             if (letter in string.punctuation):
                word = word.replace(letter, "")
        word_list[i] = word
    return word_list
      
text_list = remove_punctuation(text_list)  
```
